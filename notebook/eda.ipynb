```python
X = df["cleaned_text"]
y = df["label"]
print(X.shape)
print(y.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vectorized = vectorizer.fit_transform(X)
print(X_train_vectorized.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_vectorized, y, test_size=0.2, random_state=42)
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
print("Model trained successfully!")
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
import pickle
pickle.dump(model, open("../models/logistic_regression_model.pkl", "wb"))
pickle.dump(vectorizer, open("../models/tfidf_vectorizer.pkl", "wb"))   
print("Model saved successfully!")
```
